<h3 align="center">Exercise 7</h3> 

#### Excerise 7

Find avrage, max, min and sum of the presented array

```python
presented_array = np.arange(1,101)

print(presented_array)
```

```python
# Write code here..


mininum = presented_array.min()

maximum = presented_array.max()

avrage = presented_array.mean()

summary = presented_array.sum()

print(mininum)
print(maximum)
print(avrage)
print(summary)
```

Solution:

min --> 1

max --> 100

avrage --> 50.5

sum --> 5050
