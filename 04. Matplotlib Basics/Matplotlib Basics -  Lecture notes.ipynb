## How to make a simple function to graph

```python
import numpy as np

x = np.arange(0,100)
y = x**2


plt.plot(x,y)
plt.show()
```

___

<h3 align="center">Functional method for plotting</h3> 

#### A simple plot

```python
# Write code here..

plt.plot(x,y)

#plt.axis([0,20,0,600])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Title')

plt.show()
```

#### Making subplots

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

### Making hist-plots

```python
# Make an array of normal distributed values
z = np.random.randn(100000)
```

```python
plt.hist(z)
plt.show()
```

#### Can set the number of bins (boxes) with bins= x

```python
plt.hist(z, bins= 21)
plt.show()
```

<h3 align="center">Object oriented method for plotting </h3> 

### Create a blank canvas

```python
# Object oriented method

#Creates a figure object [think of it as an empty canvas]
fig = plt.figure()

# Add your axes to figure
axes = fig.add_axes([0.1,0.1,0.9,0.9]) #[left,bottom, width, height]

# Set labels and title
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title')
```

### Fill the canvas with a plot

```python
'''From earlier'''
#Creates a figure object [think of it as an empty canvas]
fig = plt.figure()

# Add your axes to figure
axes = fig.add_axes([0.1,0.1,0.9,0.9]) #[left,bottom, width, height]

# Set labels and title
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title')


# ------------------ This is new: --------------------


axes.plot(x,y,label='f(x) = x**2')
axes.legend()
```

### Setting colors, linewidths and linestyle

```python
'''From earlier'''
#Creates a figure object [think of it as an empty canvas]
fig = plt.figure()

# Add your axes to figure
axes = fig.add_axes([0.1,0.1,0.9,0.9]) #[left,bottom, width, height]

# Set labels and title
axes.set_xlabel('X Label')
axes.set_ylabel('y Label')
axes.set_title('Title')

# Adds a function to the canvas
axes.plot(x,y,label='f(x) = x**2')
axes.legend()

# ------------------ This is new: --------------------


# Create a new function with different color, linewidth and linestyle
axes.plot(x,1.2*x**2,color='blue', linewidth = 5, linestyle='--')
```
